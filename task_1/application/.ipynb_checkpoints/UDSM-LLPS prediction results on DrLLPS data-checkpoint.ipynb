```python
import pandas as pd
import numpy as np
```

```python
preds1_random = np.load("UDSM-LLPS-Random_predictions.npz", allow_pickle=True)
preds1_uniref = np.load("UDSM-LLPS-UniRef_predictions.npz", allow_pickle=True)
```

```python
for key in preds1_random.keys():
    print (key)
```

```python
for key in preds1_uniref.keys():
    print (key)
```

```python
# convert the prediction results into dataframes

df1_random = {
    'UniProt_IDs' : preds1_random['UniProt_IDs'].flatten(),
    'DrLLPS_labels' : preds1_random['DrLLPS_labels'].flatten(),
    'LLPS_Propensity' : preds1_random['LLPS_Propensity'][:,1].flatten()
}

df1_uniref = {
    'UniProt_IDs' : preds1_uniref['UniProt_IDs'].flatten(),
    'DrLLPS_labels' : preds1_uniref['DrLLPS_labels'].flatten(),
    'LLPS_Propensity' : preds1_uniref['LLPS_Propensity'][:,1].flatten()
}
```

```python
df1_random = pd.DataFrame(df1_random)
df1_random['case'] = 'UDSM-LLPS (Random)'
df1_uniref = pd.DataFrame(df1_uniref)
df1_uniref['case'] = 'UDSM-LLPS (UniRef)'
```
